```python
# ===============================
# COMPLETE SEABORN AI SUMMARY
# ===============================

# Import seaborn library for statistical data visualization
import seaborn as sns

# Import matplotlib.pyplot for basic plotting functions
import matplotlib.pyplot as plt

# Import pandas for handling datasets (DataFrame)
import pandas as pd


# -------------------------------
# Load Dataset
# -------------------------------

# Load the built-in "iris" dataset from seaborn
# Iris dataset contains flower measurements used in classification problems
df = sns.load_dataset("iris")


# -------------------------------
# Style setting
# -------------------------------

# Set the background style of all plots to "darkgrid"
# This improves readability
sns.set(style="darkgrid")


# -------------------------------
# 1️⃣ Histogram + KDE (Distribution Analysis)
# -------------------------------

# Create a new figure with width=6 inches and height=4 inches
plt.figure(figsize=(6,4))

# Create histogram of sepal_length
# data=df → use dataframe
# x="sepal_length" → column for x-axis
# bins=20 → divide data into 20 groups
# kde=True → show smooth density curve
# color="blue" → set bar color
sns.histplot(data=df, x="sepal_length", bins=20, kde=True, color="blue")

# Add title to the plot
plt.title("Histogram with KDE")

# Display the plot
plt.show()


# -------------------------------
# 2️⃣ Scatter Plot (Feature Relationship)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Scatter plot between sepal_length and petal_length
# hue="species" → color points by species category
# palette="Set1" → color theme
# style="species" → different marker styles
# s=100 → size of points
sns.scatterplot(data=df, x="sepal_length", y="petal_length", 
                hue="species", palette="Set1", style="species", s=100)

# Add title
plt.title("Scatter Plot")

# Show plot
plt.show()


# -------------------------------
# 3️⃣ Line Plot
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Line plot between sepal_length and petal_length
# hue="species" → separate lines for each species
sns.lineplot(data=df, x="sepal_length", y="petal_length", hue="species")

# Add title
plt.title("Line Plot")

# Show plot
plt.show()


# -------------------------------
# 4️⃣ Bar Plot (Categorical vs Numerical)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Bar plot showing average sepal_length for each species
# x="species" → categories
# y="sepal_length" → numerical value
# palette="pastel" → soft colors
sns.barplot(data=df, x="species", y="sepal_length", palette="pastel")

# Add title
plt.title("Bar Plot")

# Show plot
plt.show()


# -------------------------------
# 5️⃣ Count Plot (Categorical Count)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Count plot counts how many records exist in each species
sns.countplot(data=df, x="species", palette="Set2")

# Add title
plt.title("Count Plot")

# Show plot
plt.show()


# -------------------------------
# 6️⃣ Box Plot (Outlier Detection)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Box plot shows distribution and detects outliers
sns.boxplot(data=df, x="species", y="sepal_length", palette="Set3")

# Add title
plt.title("Box Plot")

# Show plot
plt.show()


# -------------------------------
# 7️⃣ Violin Plot (Distribution Shape)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Violin plot shows distribution shape + density
sns.violinplot(data=df, x="species", y="sepal_length", palette="muted")

# Add title
plt.title("Violin Plot")

# Show plot
plt.show()


# -------------------------------
# 8️⃣ Correlation Heatmap (Feature Importance)
# -------------------------------

# Create new figure
plt.figure(figsize=(6,4))

# Calculate correlation between numerical columns
# numeric_only=True → ignore categorical columns
corr = df.corr(numeric_only=True)

# Heatmap to visualize correlation matrix
# annot=True → show correlation values
# fmt=".2f" → show 2 decimal values
# cmap="coolwarm" → color theme
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")

# Add title
plt.title("Correlation Heatmap")

# Show plot
plt.show()


# -------------------------------
# 9️⃣ Pairplot (Complete Feature Relationship)
# -------------------------------

# Pairplot shows relationships between all numerical columns
# hue="species" → color by species
# palette="husl" → color theme
sns.pairplot(df, hue="species", palette="husl")

# Show all pairwise plots
plt.show()
```
