```python
# some oftenly used pandas functions 
import pandas as pd
import numpy as np

# Create sample DataFrame
df = pd.DataFrame({
    "Name": ["Om", "Amit", "Sara", "Riya", "John", "Om"],
    "Age": [21, 25, 22, np.nan, 30, 21],
    "City": ["Mumbai", "Delhi", "Mumbai", "Pune", "Delhi", "Mumbai"],
    "Salary": [40000, 50000, 45000, 38000, 60000, 40000]
})

# Common Pandas Functions in One Place

df.head()                      # Shows first 5 rows
df.tail()                      # Shows last 5 rows
df.sample(2)                   # Shows 2 random rows
df.shape                       # Returns (rows, columns)
df.columns                     # Shows column names
df.info()                      # Displays data types and non-null counts
df.describe()                  # Gives statistical summary of numeric columns

df["Name"]                     # Select single column
df[["Name", "Salary"]]         # Select multiple columns
df.loc[0:2]                    # Select rows by label
df.iloc[0:2]                   # Select rows by index position

df[df["Age"] > 22]             # Filter rows where Age > 22

df.isnull()                    # Shows True where values are missing
df.isnull().sum()              # Count missing values per column
df.fillna(0)                   # Replace missing values with 0

df.sort_values("Salary")       # Sort by Salary ascending
df.sort_values("Salary", ascending=False)  # Sort by Salary descending

df.groupby("City").mean()      # Group by City and calculate mean of numeric columns
df.groupby("City")["Salary"].sum()   # Group by City and sum Salary
df.groupby("City").agg(["mean", "sum"])  # Apply multiple aggregations

df["Bonus"] = df["Salary"] * 0.10     # Create new column Bonus
df.rename(columns={"Salary": "MonthlySalary"})  # Rename column
df.drop("Age", axis=1)        # Drop column Age

df["City"].unique()           # Get unique city values
df["City"].nunique()          # Count number of unique cities
df["City"].value_counts()     # Count frequency of each city

df.duplicated()               # Check duplicate rows
df.drop_duplicates()          # Remove duplicate rows

df["Salary"].apply(lambda x: x * 2)   # Apply function to each Salary value

df.to_csv("output.csv", index=False)  # Save DataFrame to CSV file
```

```python
#Merging dataframes and joining Dataframes 
df1=pd.DataFrame({'keys':['A','B','C','D'],"Values":[10,20,30,40]})
df2=pd.DataFrame({"keys":['A','B','C','D'],"Values":[20,30,40,50]})
print(f"{df1}\n\n{df2}")

print(pd.merge(df1,df2,on="keys",how="inner"))
print(pd.merge(df1,df2,on="keys",how="outer"))
print(pd.merge(df1,df2,on="keys",how="left"))
print(pd.merge(df1,df2,on="keys",how="right"))
# print(pd.merge(df1,df2,on="keys",how="Cross"))
```
