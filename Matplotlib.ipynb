```python
import pandas as pd                      # Import pandas library for data manipulation
import matplotlib.pyplot as plt          # Import matplotlib.pyplot for plotting graphs

# Read data
Data = pd.read_csv('output.csv')         # Read CSV file and store it in a DataFrame named 'Data'
print(Data)                              # Print the DataFrame to view the data


# -------------------- Plot 1: Age vs Salary --------------------

plt.figure(figsize=(8,5))                # Create a new figure (plot area) with width=8 inches and height=5 inches

plt.plot(
    Data['Age'],                         # X-axis values (Age column)
    Data['Salary'],                      # Y-axis values (Salary column)
    linestyle='--',                      # Dashed line style
    marker='o',                          # Circular markers at each data point
    color='blue'                         # Line color set to blue
)

plt.grid(True)                           # Enable grid lines on the plot
plt.xlabel("Age")                        # Label for X-axis
plt.ylabel("Salary")                     # Label for Y-axis
plt.title("Salary vs Age")               # Title of the graph
plt.show()                               # Display the plot


# -------------------- Grouping Data --------------------

Grouped_City = Data.groupby(
    'City',                              # Group data based on 'City' column
    as_index=False                       # Keep 'City' as a normal column (not index)
)['Salary'].mean()                       # Calculate mean (average) of Salary for each city

print(Grouped_City)                      # Print grouped result


# -------------------- Plot 2: Average Salary by City --------------------

plt.figure(figsize=(8,5))                # Create a new figure for second plot

plt.plot(
    Grouped_City['City'],                # X-axis values (City names)
    Grouped_City['Salary'],              # Y-axis values (Average Salary)
    marker='o'                           # Add circular markers
)

plt.xticks(rotation=45)                  # Rotate X-axis labels by 45 degrees to avoid overlap
plt.xlabel("City")                       # Label for X-axis
plt.ylabel("Average Salary")             # Label for Y-axis
plt.title("Average Salary by City")      # Title of the graph
plt.grid(True)                           # Enable grid lines
plt.show()                               # Display the second plot
```

### Histogram 
A histogram is a graph that shows how data is distributed

```python
data =[X**2 for X in range (11)]
print(data)

# creating a histogram 
plt.hist(data,bins=5)
plt.title('Histogram plot ')
plt.xlabel("X axis ")
plt.ylabel("Y axis")
plt.grid(True)


```

### Pie Chart 

```python
import matplotlib.pyplot as plt

# making a pie chart 
labels = ['A','B','C','D','E','F']
sizes = [10,20,30,25,15,10]
colors = ['green','red','yellow','orange','blue','purple']  # 6 colors
explode = (0.2,0,0,0,0,0)  # must match 6 values

# create a pie Chart 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart')

plt.show()
```

```python
# practical Example 
import pandas as pd 
import matplotlib.pyplot as plt 

Sales_data=pd.read_csv('Sale_data.csv')
Sales_data.head(10)
```

```python
Sales_data.info()
```

```python
total_sales_by_Product=Sales_data.groupby('Product')['TotalAmount'].sum()
total_sales_by_Product.plot(kind='pie',color='teal',autopct='%1.1f%%')
plt.title('Sales Summary')


```

```python
#ploting in terms of the date 
sales_trend=Sales_data.groupby('OrderDate')['TotalAmount'].sum().reset_index()
sales_trend
plt.plot(sales_trend['OrderDate'],sales_trend['TotalAmount'],color='blue')
```
